```python
import pandas as pd
```

```python
df = pd.read_csv('figures/four_peaks/ga_table.csv')
```

```python
df
```

```python
df['Best Fitness Mean'] = df['Best Fitness Mean'].apply(lambda x: f"{x/47:.3f}")
```

```python
df['Best Time Mean'] = df['Best Time Mean'].apply(lambda x: f"{x:.2f}")
```

```python
for key in ['Population Size', 'Mutation Rate',	'Best Fitness Mean', 'Best Time Mean']:
    df[key] = df[key].apply(lambda x: f"{x} & ")
```

```python
key = 'Best FEvals Mean'
df[key] = df[key].apply(lambda x: f"{x} \\\\")
```
