```python
import pandas as pd
```

```python
df = pd.read_csv('figures/four_peaks/sa_table.csv')
```

```python
df
```

```python
df['Decay Type'] = df['Decay Type'].apply(lambda x: x.replace('Decay', ''))
```

```python
df['Best Fitness Mean'] = df['Best Fitness Mean'].apply(lambda x: f"{x/47:.3f}")
```

```python
df['Best Time Mean'] = df['Best Time Mean'].apply(lambda x: f"{x:.2f}")
```

```python
for key in ['Temperature', 'Decay Type',	'Best Fitness Mean', 'Best Time Mean']:
    df[key] = df[key].apply(lambda x: f"{x} & ")
```

```python
key = 'Best FEvals Mean'
df[key] = df[key].apply(lambda x: f"{x} \\\\")
```
